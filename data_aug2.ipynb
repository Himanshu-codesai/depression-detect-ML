```python
negative_list = []
positive_list = []
for i in range(1050):
    pos = 0
    neg = 0
    row = df.iloc[i].values

    if row[0]==1 :
        pos+=1
    else:
        neg+=1

    if row[1]<=3 :
        pos+=1
    else:
        neg+=1   

    if row[2]==1 :
        pos+=1
    else:
        neg+=1
    if row[3]==1 :
        pos+=1
    else:
        neg+=1

    if row[4]==1 :
        pos+=1
    else:
        neg+=1
    if row[5]==1 :
        pos+=1
    else:
        neg+=1

    if row[6]==1 :
        pos+=1
    else:
        neg+=1
#     if row[7]>0:
#         pos+=1
#     else:
#         neg+=1

    if row[8]==1:
        pos+=1
    else:
        neg+=1
#     if row[9]>0:
#         pos+=1
#     else:
#         neg+=1
    negative_list.append(neg)
    positive_list.append(pos)
```

```python
df["positive"] = positive_list
df["negative"] = negative_list

```
