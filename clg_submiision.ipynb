```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing

import pickle
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
df = pd.read_excel('depression dataset modified.csv.xlsx')
```

```python
df
```

```python
df.shape
```

```python
label_encoder = preprocessing.LabelEncoder()

df['Result']= label_encoder.fit_transform(df['Result'])

df['Result'].unique()
```

```python
# df
```

```python
X = df.drop(columns=['With whom do you share your feelings?','Positive','Negative','How do you usually handle stress and negative emotions?','Result',])
```

```python
# X
# y
```

```python
y = df['Result']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# train - learn
# test - exam le rha
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train) 
#learn
```

```python
pred = model.predict(X_test)
#predicted ans
```

```python
# y_test
#actual ans
y_test.shape
```

```python
accuracy_score(y_test,pred)
```

```python
Pkl_Filename = "Depression_Model.pkl"  

with open(Pkl_Filename, 'wb') as file:  
    pickle.dump(model, file)
```

```python
# Load the Model back from file
with open(Pkl_Filename, 'rb') as file:  
    Pickled_LR_Model = pickle.load(file)
    
Pickled_LR_Model
```
