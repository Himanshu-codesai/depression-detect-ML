```python
import numpy as np
import pandas as pd
final = pd.read_csv(r"C:/Users/user/Downloads/finall.csv")
final
```

```python
negative_list = []
positive_list = []
for i in range(1050):
    pos = 0
    neg = 0
    row = final.iloc[i].values

    if row[0]==1 :
        pos+=1
    else:
        neg+=1

    if row[1]<=3 :
        pos+=1
    else:
        neg+=1   

    if row[2]==1 :
        pos+=1
    else:
        neg+=1
    if row[3]==1 :
        pos+=1
    else:
        neg+=1

    if row[4]==1 :
        pos+=1
    else:
        neg+=1
    if row[5]==1 :
        pos+=1
    else:
        neg+=1

    if row[6]==0 :
        pos+=1
    else:
        neg+=1
    if row[7]>0:
        pos+=1
    else:
        neg+=1

    if row[8]>0:
        pos+=1
    else:
        neg+=1
    if row[9]>0:
        pos+=1
    else:
        neg+=1
    negative_list.append(neg)
    positive_list.append(pos)
```

```python
final["positive"] = positive_list
final["negative"] = negative_list

final
```

```python
result = []
for i in positive_list:
    if i>=6 and i<=10:
        result.append("No Depression")
    elif i==4 or i==5:
        result.append("Minor Depression")
    else:
        result.append("Major Depression")
final["result"] = result
final
```

```python
X = final.drop(["result","positive", "negative"], axis=1)
y = final["result"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(X,y)
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_preds = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_preds)
```

```python
import pickle
pickle.dump(model, open('model.pkl', 'wb'))
```

```python
final.to_csv("Depression_dataset.csv", index=False)
```
